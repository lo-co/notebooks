# DMA Calculations

This notebook shows calculations associated with the data collected by Aga.  The issue seems to be that the DMA calculated diameter is different than the PSL diameter that is run through the system.  The DMA diameters that were run through here are:

* 81 $\pm$ 3 nm
* 125 $\pm$ 3 nm
* 400 $\pm$ 9 nm

```python
import matplotlib
import matplotlib.pyplot as plot
import pandas as pd
from datetime import datetime as dt

%matplotlib notebook
```

```python
# Get the data...
file = '../data/dma/20161101_16_50_31.txt'
data = pd.read_csv(file)
```

```python
# Convert date strings to an array of datetime objects....
date_obj = [dt.strptime(e, '%m/%d/%y %H:%M:%S') for e in data.Date_Time.values]
fig,ax = plot.subplots()
ax.plot(date_obj, data.DMA_Diam)
ax.set_ylabel('Diameter (nm)')
#ax.xaxis.set_tick_params()
```
